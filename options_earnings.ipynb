```python
import pandas as pd 
import numpy as np
```

```python
# Import all orders over 50 contracts on Pelton from today prior to earnings 

df = pd.read_csv(r'C:\Users\Luke\Desktop\random\pton_option.csv')
df
```

```python
# Make a column that shows the monetary value of each trade 

df['value'] = df['Qty'] * (df['Price'] * 100)
df
```

```python
# Categorize whether the option is a call or put by indexing for the last character in the option column 

calls = df[df['Option'].str.endswith('C')]
puts = df[df['Option'].str.endswith('P')]
```

```python
# Seperate the market column into two seperate columns being bid and ask 

df[['Bid','Ask']] = df['Market'].str.split('x', expand=True)
```

```python
# Convert the bid and ask columns into floats 

df['Bid']=pd.to_numeric(df.Bid)
df['Ask']=pd.to_numeric(df.Ask)
```

```python
df
```

```python
# Seperate the trades that were not traded on the bid. We are making the assumption that any midprice and orders on the ask are buy orders 

df['Trade away from Bid'] = np.where(df.Price == df.Bid, 'False', 'True')
```

```python
# Count how many buy orders we have 

df['Trade away from Bid'].value_counts()
```

```python
# Make a new column that displays whether the option traded was a call or put 

df['Type'] = np.where(df['Option'].str.endswith('C'), 'Call', 'Put')
df
```

```python
# Drop all orders that were traded on the bid 

df = df[df["Trade away from Bid"].str.contains("False")==False]
df
```

```python
# Sum up the monetary value of all the puts buying

df1 = df[df["Type"].str.contains("Call")==False]
df1
```

```python
df1['value'].sum()
```

```python
# Sum up the monetary value of all the call buying 

df2 = df[df["Type"].str.contains("Put")==False]
df2['value'].sum()
```
